#### Extracting Relevant Paper Details

```python
paper_submissions_data = []

if all_papers:
        print(f"\n--- Processing {len(all_papers)} submissions ---")

        for i in range(len(all_papers)):
            paper_in_reference = all_papers[i]

            # --- Chekcing for details and extracting replies ---
            # if hasattr(paper_in_reference, 'details') and paper_in_reference.details:
            #     # print(paper_in_reference.details)
            #     # --- Safely extract replies first ---
            #     replies_data = None # Default to None
            #     if paper_in_reference.details['replies']:
            #         replies_data = paper_in_reference.details['replies']
            #     elif paper_in_reference.details['directReplies']:
            #         # Safely get 'directReplies' from details dict
            #         replies_data = paper_in_reference.details.get('directReplies')
            #     # Ensure replies_data is a list, default to empty list if None/empty
            #     if not replies_data:
            #         replies_data = []

            paper_content = paper_in_reference.content

            if paper_content:

                # --- Build the paper_details dictionary ---
                paper_details = {
                    "paper_number": paper_in_reference.number,
                    "id": paper_in_reference.id,
                    "forum": paper_in_reference.forum,
                    "title": paper_content.get('title', {}).get('value'),
                    "authors": paper_content.get('authors', {}).get('value'),
                    "authorids": paper_content.get('authorids', {}).get('value'),
                    "keywords": paper_content.get('keywords', {}).get('value'),
                    "abstract": paper_content.get('abstract', {}).get('value'),
                    "pdf_path": paper_content.get('pdf', {}).get('value'),
                    "primary_area": paper_content.get('primary_area', {}).get('value'),
                    "venue": paper_content.get('venue', {}).get('value'),
                    "venueid": paper_content.get('venueid', {}).get('value'),
                    "code_link": paper_content.get('code', {}).get('value') or paper_content.get('code_url', {}).get('value') or paper_content.get('repository', {}).get('value'),
                    "bibtex": paper_content.get('_bibtex', {}).get('value'),
                    # Add the extracted replies data here
                    # "replies": replies_data
                }

                paper_submissions_data.append(paper_details)

            else:
                # Handle cases where paper_content is None
                print(f"Skipping paper index {i} (ID: {paper_in_reference.id}) due to missing content.")

        print(f"Successfully extracted data for {len(paper_submissions_data)} papers.")

else:
        print("No papers found for that invitation.")
```

#### Additional Formatting: URLs to Wepage, PDF and Status Filtering

```python
print(f"Starting URL generation for {len(paper_submissions_data)} entries...")

# Base URLs for construction
webpage_base_url = "https://openreview.net/forum?id="
pdf_base_url = "https://openreview.net/pdf?id="

# Keep track of any issues
missing_forum_count = 0

# Iterate through the list and update each dictionary
for paper_dict in paper_submissions_data:
    forum_id = paper_dict.get("forum") # Use .get() for safety

    if forum_id:
        # Construct the URLs
        paper_dict["webpage_url"] = webpage_base_url + forum_id
        paper_dict["pdf_url"] = pdf_base_url + forum_id
    else:
        # Handle cases where forum ID might be missing (optional)
        paper_dict["webpage_url"] = None
        paper_dict["pdf_url"] = None
        missing_forum_count += 1
        # Optionally print a warning
        # paper_id_for_warning = paper_dict.get('id', 'UNKNOWN ID')
        # print(f"Warning: Missing 'forum' ID for paper ID: {paper_id_for_warning}. URLs set to None.")

# --- Verification (Optional) ---
# Check the first entry to see if the URLs were added
if paper_submissions_data:
    print("\nExample of updated entry (first paper):")
    import pprint
    pprint.pprint(paper_submissions_data[0]) # Display the updated first dictionary
else:
    print("Data list is empty, nothing to update.")

if missing_forum_count > 0:
    print(f"\nWarning: Could not generate URLs for {missing_forum_count} entries due to missing 'forum' ID.")

print("\nFinished adding webpage_url and pdf_url keys.")
```

```python
print(f"Starting status assignment for {len(paper_submissions_data)} entries...")

# Counters for summary
status_counts = {"Accepted": 0, "Withdrawn": 0, "Rejected": 0, "Unknown": 0}
processed_count = 0

# Iterate through the list and add the 'paper_status' key
for paper_dict in paper_submissions_data:
    venue_id = paper_dict.get("venueid") # Use .get() for safety
    paper_status = "Unknown" # Default status

    if venue_id and isinstance(venue_id, str): # Check it exists and is a string
        if "Withdrawn_Submission" in venue_id:
            paper_status = "Withdrawn"
        elif "Rejected_Submission" in venue_id:
            # Could check for Desk Rejected as it's distinct
            # if "Desk_Rejected_Submission" in venue_id:
            #    paper_status = "Desk Rejected"
            # else:
                paper_status = "Rejected"
        else:
            # If not explicitly withdrawn or rejected/desk-rejected, assume Accepted
            paper_status = "Accepted"

        paper_dict["paper_status"] = paper_status
        status_counts[paper_status] += 1
        processed_count += 1
    else:
        # Handle cases where venueid is missing or not a string
        paper_dict["paper_status"] = "Unknown"
        status_counts["Unknown"] += 1

# --- Verification and Summary ---
print(f"\nFinished assigning status for {processed_count} entries.")
print("Status breakdown:")
for status, count in status_counts.items():
    if count > 0: # Only print statuses that occurred
        print(f"- {status}: {count}")

# Optional: Check the first entry to see if the status was added
if paper_submissions_data:
    print("\nExample of updated entry (first paper):")
    import pprint
    # Limit printing depth for clarity if needed
    pprint.pprint(paper_submissions_data[0], depth=2)
else:
    print("Data list is empty, nothing to update.")
```

#### Download CSV

```python
import csv

output_csv_filename = f'{safe_venue_id_part}_summary.csv' # Define output filename

print(f"Preparing to write data to {output_csv_filename}...")

# --- Check if data exists ---
if not paper_submissions_data:
    print("Variable 'paper_submissions_data' is empty. No CSV file written.")
else:
    # --- Define CSV Columns (Fieldnames) ---
    # Extract keys from the first paper dictionary
    fieldnames = [key for key in paper_submissions_data[0].keys()]
    print(f"CSV columns will be: {fieldnames}")

    # --- Write Output to CSV File ---
    try:
        with open(output_csv_filename, 'w', newline='', encoding='utf-8') as csvfile:
            writer = csv.DictWriter(csvfile, fieldnames=fieldnames)

            # Write the header row
            writer.writeheader()

            # Write the data rows
            processed_count = 0
            for paper_dict in paper_submissions_data:
                # Create a row dictionary containing only the specified fieldnames
                row_data = {}
                for key in fieldnames:
                    value = paper_dict.get(key) # Use .get() for safety if keys might vary slightly

                    # Handle lists (e.g., authors, keywords) by joining them
                    if isinstance(value, list):
                        # Join list items into a single string, handling non-string items
                        row_data[key] = '; '.join(map(str, value))
                    elif value is None:
                        row_data[key] = '' # Represent None as empty string in CSV
                    else:
                        # Ensure value is a string for CSV writing
                        row_data[key] = str(value)

                writer.writerow(row_data)
                processed_count += 1

        print(f"Finished writing {processed_count} rows to {output_csv_filename}")

    except Exception as e:
        print(f"\nAn error occurred during CSV writing: {e}")
        import traceback
        traceback.print_exc()
```
